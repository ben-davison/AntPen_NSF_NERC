```python
import xarray as xr
```

```python
data_loc = '/home/jovyan/data/RACMO/t2m/'

start_string = 't2m.KNMI-'
end_string_1 = '.XPEN055.ERAINx_RACMO2.3p2.DD.nc'
end_string_2 = '.XPEN055.ERA5-3H_RACMO2.3p2.3H.nc'
```

```python
filename_1 = data_loc + start_string + '1979' + end_string_1
filename_2 = data_loc + start_string + '1981' + end_string_1
filename_3 = data_loc + start_string + '1991' + end_string_1
filename_4 = data_loc + start_string + '2001' + end_string_1
filename_5 = data_loc + start_string + '2011' + end_string_1
filename_6 = data_loc + start_string + '2016' + end_string_2
filename_7 = data_loc + start_string + '2021' + end_string_2
```

```python
t2m_6 = xr.open_dataset(filename_6,chunks={'time':10})
t2m_6 = t2m_6.resample(time='D').mean()
```

```python
t2m_7 = xr.open_dataset(filename_7,chunks={'time':10})
t2m_7 = t2m_7.resample(time='D').mean()
```

```python
t2m_1 = xr.open_dataset(filename_1).drop(['time_bnds'])
t2m_2 = xr.open_dataset(filename_2).drop(['time_bnds'])
t2m_3 = xr.open_dataset(filename_3).drop(['time_bnds'])
t2m_4 = xr.open_dataset(filename_4).drop(['time_bnds'])
t2m_5 = xr.open_dataset(filename_5).drop(['time_bnds'])
```

```python
t2m = xr.concat([
    t2m_1,
    t2m_2,
    t2m_3,
    t2m_4,
    t2m_5,
    t2m_6,
    t2m_7,
],dim='time').drop_duplicates('time',keep='first').squeeze()
```
